```python
import pandas as pd
import altair as alt
```

```python
car_sales = pd.read_csv ("Norway_car_sales.csv")
```

```python
brush = alt.selection_interval()
```

```python
diesel_sales = alt.Chart(car_sales).mark_bar().encode(
    x='Year:N',
    y='Quantity_Diesel:Q',
    color=alt.condition(brush, 'Month:N', alt.value('lightgray'))
).add_selection(
    brush
)

diesel_sales
```

The amount of diesel cars sold each year is distributed fairly evenly across each month. Since 2011, the amount of diesel cars sold has declined year on year.

```python
electric_sales = alt.Chart(car_sales).mark_bar().encode(
    x='Year:N',
    y='Quantity_Electric:Q',
    color=alt.condition(brush, 'Month:N', alt.value('lightgray'))
).add_selection(
    brush
)

electric_sales
```

Electric car sales have increased year on year from 2011 to 2015. Although they have then fallen in 2016, they haven't done so by much. Sales by month are not even, with some months repeatedly having higher electric car sales and some much lower. This can be seen when comparing March with July for 2014, 2015 and 2016. In 2013, the months with the highest electric car sales are August to December. This could possibly be explained by govenment incentives as the quantity of electric car sales more than doubles in the following year.

```python
diesel_sales & electric_sales
```

When looking at both bar charts together, it is clear to see that the increase in electric car sales coincides with the decrease in diesel car sales. 

```python
average_CO2 = alt.Chart(car_sales).mark_point().encode(
    x='Year:N',
    y='Avg_CO2:Q',
    color=alt.condition(brush, 'Year:N', alt.value('lightgray'))
).add_selection(
    brush
)

average_CO2
```

In the graph above, it is clear that average CO2 levels have decreased steadily from 2007 to 2017.

```python
diesel_sales & average_CO2
```

After comparing the Average CO2 graph with the diesel sales chart, it is clear to see that the decreasing CO2 levels coincides with the decreasing diesel car sales.

I chose to use bar charts for the Diesel sales and Electric sales as this showed the total sales for the year clearly. I then chose to colour the bars by monthly sales so it was clear how much of the total was sold each month. For the average CO2 data, I decided that a scatter plot graph would be best as it showed the decline in CO2 from 2007 to 2017. I would have considered a line graph if there had been just one amount for each year but the data is provided monthly in the CSV file. For the scatter plot graph, I chose to colour by year because I didn't want the focus to be on the amount each month.
